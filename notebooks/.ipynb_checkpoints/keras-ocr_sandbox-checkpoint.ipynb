```python
import matplotlib.pyplot as plt
import keras_ocr
import os
%matplotlib inline
import sqlalchemy
import pandas as pd
import numpy as np
```

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
#fpath = os.path.join('..', 'data', 'processed', 'CPO', 'f5844df2d7d609c7ae385e446b19c02d.png')
fpath = os.path.join('..', 'data', 'processed', 'CHS', 'cheerios-giant-size2.png')
image = keras_ocr.tools.read(fpath)

predictions = pipeline.recognize([image])[0]
ocr_words = set([pred[0] for pred in predictions]) # make set for jaccard sim

# Plot the predictions
keras_ocr.tools.drawAnnotations(image=image, predictions=predictions);
```

```python
# Get cereals from DB
passwd = open('../db_password','r').readlines()[0].split()[0]
username = open('../db_user','r').readlines()[0].split()[0]
dbname = 'cereals'

# Connect to database
db = sqlalchemy.create_engine(f'mysql+pymysql://{username}:{passwd}@localhost/{dbname}')
conn = db.connect()
query = 'select cereal_name, company from cereals'
df = pd.read_sql(query, conn)
# add jaccard column
df["jaccard"] = np.nan
```

```python
def get_jaccard_sim(set1, set2):
    intersect = set1.intersection(set2)
    return len(intersect) / (len(set1) + len(set2) - len(intersect))

for ix, row in df.iterrows():
    cereal = row['cereal_name'] + " " + row['company'] # use both cereal and company name
    cereal = cereal.lower()             # lowercase
    cereal = cereal.replace("'","")     # remove apostrophes
    cereal = cereal.replace("-", " ")   # replace any hyphens with white space
    cereal = set(cereal.split())        # split on white space and make set
    
    jaccard = get_jaccard_sim(ocr_words, cereal)
    df.loc[ix, "jaccard"] = jaccard
    
df.sort_values(by=['jaccard'], ascending=False)    
```
