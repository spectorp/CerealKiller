```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

obj = cv2.imread(os.path.join('..', 'data', 'raw', 'CHX', '51JXYOJn6SL._SY445_.jpg'))
b,g,r = cv2.split(obj)           # get b, g, r
obj = cv2.merge([r,g,b])     # switch it to r, g, b

im = cv2.imread(os.path.join('..', 'data', 'raw', 'shelves', '6014025_031420-kfsn-6p-local-business-coronavirus-vid.jpg'))
b,g,r = cv2.split(im)           # get b, g, r
im = cv2.merge([r,g,b])     # switch it to r, g, b

mask = 255 * np.ones(obj.shape, obj.dtype)
```

```python
# The location of the center of the src in the dst
width, height, channels = im.shape
center = (int(height/2), int(width/2))

# Seamlessly clone src into dst and put the results in output
normal_clone = cv2.seamlessClone(obj, im, mask, center, cv2.NORMAL_CLONE)

Image.fromarray(normal_clone, 'RGB')
```
